```
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```
rank = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/cwurData.csv")
merged = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/merged_data.csv")
```

```
# data wrangling 
rank = rank[rank['country'] == 'USA']
rank = rank[['institution','national_rank']]
rank = rank.drop_duplicates(subset='institution')
```

```
rank.sort_values(by='national_rank')
```

```
#full = merged.merge(rank, left_on='INSTNM', right_on='institution', how='left')
full = pd.merge(merged, rank, left_on='INSTNM', right_on='institution', how='left')
```

```
# drop duplicate column & rearrange
full = full.drop(columns=['institution'])
full = full[ ['national_rank'] + [ col for col in full.columns if col != 'national_rank' ] ]
```

```
# sort by rank
full = full.sort_values(by='national_rank').reset_index(drop=True)
full
```

```
full[full['ADDR'] == " "]['INSTNM'].drop_duplicates().tolist()
```

```
# handle some empty address
full.loc[full['INSTNM'] == 'Stanford University','ADDR'] = "450 Serra Mall"
full.loc[full['INSTNM'] == 'University of Florida','ADDR'] = "Tigert Hall"
full.loc[full['INSTNM'] == 'Michigan State University'] = "220 Trowbridge Rd"
full.loc[full['INSTNM'] == 'Baylor University'] = "500 Speight Ave"
full.loc[full['INSTNM'] == 'Auburn University'] = "1161 West Samford Avenue"
full.loc[full['INSTNM'] == 'University of Maryland-College Park'] = "College Park"
full.loc[full['INSTNM'] == 'Hiram College'] = "11715 Garfield Road"
full.loc[full['INSTNM'] == 'Abilene Christian University'] = "1600 Campus Ct"
```

```
full.to_csv('ranked_merged_data.csv')
```
